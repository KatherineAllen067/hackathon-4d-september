```python
local_in_review = []

for text in df_yelp['text']:
    if "local" in text:
        local_in_review.append(1)
    elif "Local" in text:
        local_in_review.append(1)
    elif "locals" in text:
        local_in_review.append(1)
    elif "Locals" in text:
        local_in_review.append(1)
    else:
        local_in_review.append(0)
```

```python
# check length
len(local_in_review)
```

```python
df_local = pd.DataFrame({'local_in_review': local_in_review})

#check
df_local
```

```python
df_yelp = pd.concat([df_yelp, df_local], axis=1)
```

```python
# check
df_yelp
```

```python
# check how many reviews have local in them
df_yelp['local_in_review'].value_counts()
```

```python
# converting the date object to a datetime type
df_yelp['date'] = pd.to_datetime(df_yelp['date'], infer_datetime_format=True)

# converting the date column into year, month, day columns
df_yelp['review_year'] = df_yelp['date'].dt.year
df_yelp['review_month'] = df_yelp['date'].dt.month
df_yelp['review_day'] = df_yelp['date'].dt.day

df_yelp
```

```python
#drop the 'date' column
df_yelp.drop(['date'], axis=1, inplace=True)
```

```python
# check if attributes remain the same for restaurants
print(df_yelp['attributes'][0])
print(df_yelp['attributes'][1])
print(df_yelp['attributes'][2])
print(df_yelp['attributes'][3])
```
